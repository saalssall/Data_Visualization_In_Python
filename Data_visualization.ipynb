```python
# Import the matplotlib.pyplot submodule and name it plt
import matplotlib.pyplot as plt
import pandas as pd

titanic_df = pd.read_csv('Data/titanic.csv')
print(titanic_df.columns.tolist())
```

```python
# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots()

# Plot Age distribution for males
males = titanic_df[titanic_df['Sex'] == 'male']
females = titanic_df[titanic_df['Sex'] == 'female']

# Plot Fare against PassengerId for males and females
ax.plot(males['Age'], males['Fare'], label='Male')
ax.plot(females['Age'], females['Fare'], label='Female')
plt.show()
```

```python
fig, ax = plt.subplots()

# Import data
churn_df = pd.read_csv('Data/telecom_churn.csv')

# Subset by churn status
churned = churn_df[churn_df['Churn'] == 'Yes']
not_churned = churn_df[churn_df['Churn'] == 'No']

# Sample 15 observations from each group
churned_sample = churned.sample(min(15, len(churned)), random_state=321)
not_churned_sample = not_churned.sample(min(15, len(not_churned)), random_state=321)

# Plot churned data
ax.plot(churned_sample['tenure'], churned_sample['MonthlyCharges'], color='b', marker='o', linestyle='--', label='Churned')
# Plot not churned data
ax.plot(not_churned_sample['tenure'], not_churned_sample['MonthlyCharges'], color='r', marker='v', linestyle='--', label='Not Churned')

plt.legend()
ax.set_xlabel('Tenure (Months)')
ax.set_ylabel('Monthly Charges (USD)')
ax.set_title('Tenure vs Monthly Charges by Churn Status (n=15)')
plt.show()
```

```python
# Create a Figure and an array of subplots with 2 rows and 2 columns
fig, ax = plt.subplots(2, 2)

# Top left: tenure vs MonthlyCharges for churned
ax[0, 0].plot(churned_sample['tenure'], churned_sample['MonthlyCharges'], color='b')
ax[0, 0].set_title('Churned: Tenure vs Monthly Charges')

# Top right: tenure vs TotalCharges for churned
ax[0, 1].plot(churned_sample['tenure'], pd.to_numeric(churned_sample['TotalCharges'], errors='coerce'), color='b')
ax[0, 1].set_title('Churned: Tenure vs Total Charges')

# Bottom left: tenure vs MonthlyCharges for not churned
ax[1, 0].plot(not_churned_sample['tenure'], not_churned_sample['MonthlyCharges'], color='r')
ax[1, 0].set_title('Not Churned: Tenure vs Monthly Charges')

# Bottom right: tenure vs TotalCharges for not churned
ax[1, 1].plot(not_churned_sample['tenure'], pd.to_numeric(not_churned_sample['TotalCharges'], errors='coerce'), color='r')
ax[1, 1].set_title('Not Churned: Tenure vs Total Charges')

plt.tight_layout()
plt.show()
```

```python
# Create a figure and an array of axes: 2 rows, 1 column with shared y axis
fig, ax = plt.subplots(2, 1, sharey=True)

# Plot churned data in the top axes
ax[0].plot(churned_sample['tenure'], churned_sample['MonthlyCharges'], color='b')
ax[0].plot(churned_sample['tenure'], pd.to_numeric(churned_sample['TotalCharges'], errors='coerce'), color='b', linestyle='--')
ax[0].set_title('Churned')

# Plot not churned data in the bottom axes
ax[1].plot(not_churned_sample['tenure'], not_churned_sample['MonthlyCharges'], color='r')
ax[1].plot(not_churned_sample['tenure'], pd.to_numeric(not_churned_sample['TotalCharges'], errors='coerce'), color='r', linestyle='--')
ax[1].set_title('Not Churned')

plt.tight_layout()
plt.show()
```
